```python
n_customers = 2000

customer_id = [f"C{str(i).zfill(4)}" for i in range(1, n_customers+1)]

gender = np.random.choice(["Male", "Female", "Other"], size=n_customers, p=[0.48, 0.48, 0.04])
age = np.random.randint(18, 81, size=n_customers)
country = np.random.choice(["Finland", "Sweden", "Norway", "Denmark", "Germany"], size=n_customers, p=[0.35, 0.2, 0.15, 0.15, 0.15])

tenure = np.random.randint(0, 121, size=n_customers)

contract_type = []
for t in tenure:
    if t < 6:
        contract_type.append(np.random.choice(["Month-to-month", "One year"], p=[0.8, 0.2]))
    elif t < 24:
        contract_type.append(np.random.choice(["Month-to-month", "One year", "Two year"], p=[0.5, 0.3, 0.2]))
    else:
        contract_type.append(np.random.choice(["Month-to-month", "One year", "Two year"], p=[0.3, 0.3, 0.4]))

payment_method = np.random.choice(
    ["Credit card", "Bank transfer", "Direct debit", "Electronic invoice"],
    size=n_customers,
    p=[0.3, 0.25, 0.25, 0.2]
)

internet_service = np.random.choice(["DSL", "Fiber", "Mobile", "None"], size=n_customers, p=[0.3, 0.4, 0.25, 0.05])
has_phone = np.random.choice(["Yes", "No"], size=n_customers, p=[0.8, 0.2])
addon_streaming = np.random.choice(["Yes", "No"], size=n_customers, p=[0.4, 0.6])

base_price = np.random.uniform(20, 60, size=n_customers)
fiber_addon = np.where(internet_service == "Fiber", np.random.uniform(10, 40, size=n_customers), 0)
mobile_discount = np.where(internet_service == "Mobile", -np.random.uniform(0, 10, size=n_customers), 0)
streaming_addon = np.where(addon_streaming == "Yes", np.random.uniform(5, 15, size=n_customers), 0)

monthly_charges = base_price + fiber_addon + mobile_discount + streaming_addon
monthly_charges = np.round(np.clip(monthly_charges, 15, None), 2)

total_charges = monthly_charges * tenure + np.random.normal(0, 20, size=n_customers)
total_charges = np.round(np.clip(total_charges, 0, None), 2)

support_tickets = np.random.poisson(lam=1.5, size=n_customers)
support_tickets = np.clip(support_tickets, 0, 10)

satisfaction = (
    4.5
    - (monthly_charges - monthly_charges.mean()) / monthly_charges.std() * 0.2
    - support_tickets * 0.2
    + (tenure / 120) * 0.5
)
satisfaction = np.clip(np.round(satisfaction), 1, 5)

is_active_app_user = np.where(np.random.rand(n_customers) < 0.6, "Yes", "No")

churn_prob = (
    0.25
    + np.where(np.array(contract_type) == "Month-to-month", 0.15, 0)
    + np.where(tenure < 6, 0.15, 0)
    + np.where(satisfaction <= 2, 0.2, 0)
    + np.where(support_tickets >= 4, 0.1, 0)
    + (monthly_charges - monthly_charges.mean()) / monthly_charges.std() * 0.05
    - np.where(is_active_app_user == "Yes", 0.05, 0)
)

churn_prob = np.clip(churn_prob, 0.01, 0.8)
churn = np.where(np.random.rand(n_customers) < churn_prob, 1, 0)

churn_reason = []
for c, s in zip(churn, satisfaction):
    if c == 0:
        churn_reason.append("")
    else:
        if s <= 2:
            churn_reason.append(np.random.choice(["Service quality", "Price"]))
        else:
            churn_reason.append(np.random.choice(["Competitor", "Moving", "Other"]))
```
